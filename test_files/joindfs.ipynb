```python
import pandas as pd
import os
import re
```

Join files

```python
re_assn = '.*_assignments'
re_courses = '.*_courses'
re_submissions = '.*_submissions'
# ----- Join assignments --------------------------------------------
assignment_fnames = []
# Get assignment files
for fname in os.listdir():
    if re.match(re_assn, fname):
        assignment_fnames.append(fname)
# Load assignments as dataframes   
assignments= pd.read_parquet(assignment_fnames[0])
for fname in assignment_fnames[1:]:
    assignments = pd.concat([assignments,pd.read_parquet(fname)])
# ----- Join courses ----------------------------------------------
course_fnames = []
# Get courses files
for fname in os.listdir():
    if re.match(re_courses, fname):
        course_fnames.append(fname)
# Load courses as dataframes   
courses = pd.read_parquet(course_fnames[0])
for fname in course_fnames[1:]:
    courses = pd.concat([courses,pd.read_parquet(fname)])
# ----- Join submissions ----------------------------------------------
submission_fnames = []
# Get submissions files
for fname in os.listdir():
    if re.match(re_submissions, fname):
        submission_fnames.append(fname)
# Load submissions as dataframes   
submissions = pd.read_parquet(submission_fnames[0])
for fname in submission_fnames[1:]:
    submissions = pd.concat([submissions,pd.read_parquet(fname)])
```

Output files

```python
# Save to a Parquet file
submissions.to_parquet('submissions', index=False)
courses.to_parquet('courses', index=False)
assignments.to_parquet('assignments', index=False)
```
